```python
import tensorflow as tf
saved_model_dir = "/content/drive/MyDrive/CNN model2/model.h5"
model = tf.keras.models.load_model(saved_model_dir)
```

```python
##Extracting the files and saving the names for testing the saved model
import os
# Define the path to the 'Healthy' folder within the main_data_dir
main_data_dir=r'/content/drive/MyDrive/CNN model2/Test Data'
healthy_folder_path = os.path.join(main_data_dir, r'/content/drive/MyDrive/CNN model2/Test Data/Healthy')
miner_folder_path= os.path.join(main_data_dir, r'/content/drive/MyDrive/CNN model2/Test Data/Miner')
def data_path(folder_path):
    # Get a list of files in the 'Healthy' folder
    healthy_files = os.listdir(healthy_folder_path)
    files=[]
    #Iterate through the files
    for file_name in healthy_files:
        file_path = os.path.join(healthy_folder_path, file_name)
        files.append(file_path)
    print('filed saved successfully')
    return files
healthy_files=data_path(healthy_folder_path)
miner_files=data_path(miner_folder_path)
```

```python
healthy_files[:2]
```

```python
def test_data(image_path):
  import cv2
  import numpy as np
  img = cv2.imread(image_path)
  resize = tf.image.resize(img, (256,256))
  yhat = model.predict(np.expand_dims(resize/255, 0))
  print('yhat value',yhat)
  if yhat > 0.5:
    #print(f'Miner')
    return 'Miner'
  else:
      #print(f'Healthy')
      return 'Healthy'
```

```python
correct=0
wrong=0
wrong_files=[]
for image in healthy_files:
  output=test_data(image)
  if output=='Healthy':
    correct+=1
  else:
    wrong+=1
    wrong_files.append(image)
```

```python
print('correctly classified count',correct)
```

```python
correct=0
wrong=0
wrong_files=[]
for image in miner_files:
  output=test_data(image)
  if output=='Miner':
    correct+=1
  else:
    wrong+=1
    wrong_files.append(image)
print('correctly classified',correct)
print('wrongly classified')
if wrong_files:
  print('file names of wrongly classified',wrong_files)
```
